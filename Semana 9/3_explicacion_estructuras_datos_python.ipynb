### üß† Explicaci√≥n: Tuplas, Listas y Diccionarios en Python

```
# =========================
# LISTAS EN PYTHON
# =========================
# Las listas son colecciones ordenadas y modificables (mutables). Se definen con corchetes []

mi_lista = [10, 20, 30, 40]
print("Lista original:", mi_lista)

# Podemos acceder a los elementos por √≠ndice
print("Primer elemento:", mi_lista[0])

# Agregar un elemento
mi_lista.append(50)
print("Despu√©s de agregar 50:", mi_lista)

# Modificar un elemento
mi_lista[1] = 25
print("Despu√©s de modificar el segundo elemento:", mi_lista)

# Eliminar un elemento
mi_lista.remove(30)
print("Despu√©s de eliminar 30:", mi_lista)

# Longitud de la lista
print("Longitud:", len(mi_lista))
```

```
# =========================
# TUPLAS EN PYTHON
# =========================
# Las tuplas son colecciones ordenadas pero INMUTABLES. Se definen con par√©ntesis ()

mi_tupla = (100, 200, 300)
print("\nTupla original:", mi_tupla)

# Acceso por √≠ndice
print("Segundo elemento:", mi_tupla[1])

# No se pueden modificar ni a√±adir elementos (esto causar√≠a error)
# mi_tupla[0] = 999  # Descomentar esto generar√° un error

# Las tuplas se usan cuando los datos no deben cambiar
# Tambi√©n pueden usarse para retornar m√∫ltiples valores de una funci√≥n
```

```
# =========================
# DICCIONARIOS EN PYTHON
# =========================
# Los diccionarios son colecciones desordenadas de pares clave:valor. Se definen con llaves {}

mi_diccionario = {
    "nombre": "Ana",
    "edad": 30,
    "ciudad": "Quito"
}

print("\nDiccionario original:", mi_diccionario)

# Acceder a un valor por su clave
print("Nombre:", mi_diccionario["nombre"])

# Agregar una nueva clave:valor
mi_diccionario["profesion"] = "Ingeniera"
print("Despu√©s de agregar profesi√≥n:", mi_diccionario)

# Modificar un valor
mi_diccionario["edad"] = 31
print("Despu√©s de modificar edad:", mi_diccionario)

# Eliminar una clave
del mi_diccionario["ciudad"]
print("Despu√©s de eliminar ciudad:", mi_diccionario)

# Mostrar todas las claves
print("Claves:", list(mi_diccionario.keys()))

# Mostrar todos los valores
print("Valores:", list(mi_diccionario.values()))
```

### ‚úÖ Resumen

| Estructura   | Ordenada | Mutable | Sintaxis        | Uso com√∫n                                     |
|--------------|----------|---------|------------------|-----------------------------------------------|
| Lista        | S√≠       | S√≠      | `[ ]`           | Conjuntos de elementos que pueden cambiar     |
| Tupla        | S√≠       | No      | `( )`           | Datos que no deben cambiar                    |
| Diccionario  | No       | S√≠      | `{ clave:valor }`| Asociar claves con valores (tipo JSON)        |
