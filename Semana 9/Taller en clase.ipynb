# Trabajo en Clase: “Sistema Inteligente de Gestión de Clientes”
## Objetivos del ejercicio:

* Aplicar algoritmos de ordenamiento y búsqueda para manipular información realista.
* Comprender cómo ordenar listas de tuplas por diferentes criterios.
* Evaluar la eficiencia de la búsqueda binaria comparada con la búsqueda lineal.
* Analizar críticamente cuál algoritmo es más adecuado según el contexto.

## Enunciado del problema
Una empresa de servicios quiere automatizar la gestión de su base de datos de clientes. Para ello, te han solicitado crear un pequeño sistema en Python que permita:

1. Ordenar clientes por nombre alfabéticamente.
2. Ordenar clientes por saldo adeudado de menor a mayor.
3. Buscar un cliente por su nombre para verificar si se encuentra en la base.

## Requisitos funcionales
### Parte 1: Datos y ordenamiento

* Crea una lista de al menos 10 clientes representados como tuplas:

    Formato: (nombre_cliente, saldo)
    clientes = [("Carlos", 100.0), ("Ana", 50.5), ...]

* Implementa dos algoritmos de ordenamiento desde cero:
    - Insertion Sort para ordenar por nombre.
    - Selection Sort para ordenar por saldo.

### Parte 2: Búsqueda

* Solicita al usuario un nombre de cliente.
* Usa búsqueda binaria para verificar si el cliente está en la base de datos.
* Muestra:
    * Si el cliente fue encontrado.
    * En qué posición.
    * Cuántas comparaciones se realizaron.

### Parte 3: Preguntas de reflexión (escritas en celda Markdown)

* ¿Por qué es necesario ordenar antes de realizar la búsqueda binaria?
* ¿Qué pasaría si usamos búsqueda binaria sin ordenar primero?
* ¿Qué ventajas viste entre la búsqueda binaria y la búsqueda lineal?
* ¿Cuál de los dos ordenamientos te pareció más intuitivo de implementar y por qué?

### Extra:
* Permite al usuario ingresar un nuevo cliente (nombre + saldo).
* Usa búsqueda lineal si el usuario elige no ordenar la lista.
