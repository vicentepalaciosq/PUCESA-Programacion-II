# COLAS EN PYTHON - PARTE TEÓRICA

Una **cola** (queue) es una estructura de datos lineal que sigue el principio FIFO (First In, First Out), es decir, el primer elemento en entrar es el primero en salir.

### Ejemplos de uso de colas en la vida real:
- Cola de personas esperando en un banco.
- Cola de impresión de documentos.
- Cola de tareas a ejecutar por el sistema operativo.

## ¿Por qué estudiar colas?

Las colas son fundamentales en computación porque permiten gestionar procesos, tareas y solicitudes de forma ordenada, garantizando que los recursos se asignen en el orden correcto.

Se utilizan, por ejemplo, en:
- Sistemas de atención al cliente.
- Programación de procesos.
- Transmisión de datos.
- Buffers de entrada/salida.
- Simulaciones de sistemas reales.

## Operaciones básicas de una cola

| Operación | Descripción |
|-----------|-------------|
| `enqueue()` | Agrega un elemento al final de la cola |
| `dequeue()` | Elimina el elemento al frente de la cola |
| `is_empty()` | Verifica si la cola está vacía |
| `peek()` o `front()` | Muestra el elemento al frente sin eliminarlo |
| `size()` | Devuelve el tamaño actual de la cola |

## Principio FIFO (First In, First Out)

Este principio indica que el primer elemento que entra a la cola será el primero en salir.

### Ejemplo gráfico:

```text
Inicio: [ ]
enqueue(A) -> [A]
enqueue(B) -> [A, B]
enqueue(C) -> [A, B, C]
dequeue()  -> sale A -> [B, C]
```

Se observa cómo los elementos son atendidos en el mismo orden en que fueron agregados.

## Implementaciones de colas en Python

En Python existen varias formas de implementar colas:
- Usando **listas (list)**
- Usando **collections.deque** (más eficiente)
- Usando módulos especializados como **queue.Queue** (para programación concurrente)
- Implementando una clase propia de **Cola**
